```python
import pandas as pd
import os
```

```python
path = './Sales_Data'
files = [file for file in os.listdir(path) if not file.startswith('.')]
df = pd.DataFrame()
for file in files:
    df1 = pd.read_csv(path+'/'+file)
    df = pd.concat([df,df1])
```

```python
df
```

```python
df.to_csv('combined.csv',index=False)
```

```python
df_combined = pd.read_csv('combined.csv')
df_combined
```

```python
nan_df = df_combined[df_combined.isna().any(axis=1)]
display(nan_df)
```

```python
df_combined = df_combined.dropna(how='all')
```

```python
df_combined = df_combined.convert_dtypes()
```

```python
# Extract month from order Date
df_combined['month'] = df_combined['Order Date'].str[0:2]
```

```python
df_combined = df_combined[df_combined['Order Date'].str[0:2] != 'Or']
df_combined
```

```python
df_combined['month'] = df_combined['month'].astype('int16')
```

```python
df_combined['Price Each'] = df_combined['Price Each'].astype('float32')
```

```python
df_combined['Quantity Ordered'] = df_combined['Quantity Ordered'].astype('int32')
```

```python
df_combined.dtypes
```

```python
# Add a new column "city"
df_combined['City'] = df_combined['Purchase Address'].apply(lambda x: x.split(',')[1].strip(' ')+' '+x.split(',')[2].strip(' ').split(' ')[0])
df_combined
```

#### What products are most often sold together?

```python
df = df_combined[df_combined['Order ID'].duplicated(keep=False)]
df
```

```python
df['All Products'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df_new=df[['Order ID','All Products']].drop_duplicates()
df_new
```

```python
from itertools import combinations
from collections import Counter

count  = Counter()

for row in df_new['All Products']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key,value in count.most_common(10):
    print(key, value)
    
```

#### What product was sold the most. and why do you think it was sold the most?

```python
df_product = df.groupby('Product')
Q_ordered = df_product.sum()['Quantity Ordered']
```

```python
Q_ordered
```

```python
keys = [pair for pair, df in df_product]
plt.bar(keys,Q_ordered)
plt.xticks(keys,rotation='vertical',size=8)
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
# Hypothesis that USB charging cable and lightning charging cabls are sold the most because their prices are. low and they are most used
```

```python
prices = df.groupby('Product').mean()['Price Each']
prices

fig,ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(keys,Q_ordered)
ax2.plot(keys,prices,color='b')


ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered',color='g')
ax2.set_ylabel('Price($)',color='b')
ax1.set_xticklabels(keys,rotation='vertical',size=8)

plt.show()
```
